#### 1. **What is the dependence in Europe on fossil fuels (oil, coal, and natural gas)?**
   Europe has historically been highly dependent on fossil fuels for its energy needs, with oil, coal, and natural gas serving as the cornerstone of its industrial and economic development. However, this dependence has been subject to change, driven by factors such as environmental concerns, geopolitical events, and the evolving energy market. The region's fossil fuel consumption has fluctuated over time, with notable shifts in the energy mix, particularly in response to climate change targets and the need for energy security.

#### 2. **What is the current share of fossil fuels in Europe's total energy consumption?**
   As of the latest data, fossil fuels continue to make up a significant portion of Europe’s total energy consumption. Despite growing investments in renewable energy sources like solar, wind, and biofuels, oil, coal, and natural gas still account for a substantial share of the energy mix. The share of fossil fuels varies by country and sector, with some nations more dependent on specific fuels than others. This ongoing reliance is both a challenge and an area of focus for Europe's energy policy, which aims to reduce its carbon footprint and transition towards cleaner alternatives.

#### 3. **How does the reliance on fossil fuels vary across different European countries?**
   The degree of dependence on fossil fuels differs significantly across Europe. Some countries, such as Poland and Germany, have historically been major consumers of coal, while others, like Norway and the Netherlands, have seen a stronger shift toward natural gas. The variation is driven by factors including domestic resource availability, economic structure, and national energy policies. For example, countries like France have invested heavily in nuclear energy, reducing their fossil fuel consumption, while others have increasingly relied on natural gas imports, particularly from Russia and the Middle East.

#### 4. **How has this developed over time?**
   Europe's fossil fuel dependence has evolved significantly over the past century. During the early stages of industrialization, coal was the primary energy source, with countries like the UK, Germany, and Poland being major producers. In the latter half of the 20th century, oil and natural gas began to play a larger role, with imports from the Middle East and later from Russia shaping the energy landscape. Major geopolitical events, such as the **1973 oil crisis**, the **fall of the Soviet Union**, and the **Russia-Ukraine war**, have had profound impacts on Europe's fossil fuel consumption patterns. These events, combined with the EU's environmental goals, have driven a gradual shift toward renewable energy sources and increased diversification of energy imports.

#### 5. **What role have major geopolitical events (e.g., the 1973 oil crisis, the fall of the Soviet Union, etc.) played in shaping Europe's fossil fuel dependence?**
   Geopolitical events have played a crucial role in shaping Europe’s fossil fuel dependence. The **1973 oil crisis**, caused by the OPEC oil embargo, highlighted Europe's vulnerability to external oil supplies, sparking a drive to diversify energy sources and reduce reliance on oil imports. Similarly, the **fall of the Soviet Union** opened up new opportunities for energy trade with Russia, resulting in increased natural gas imports. The **Russia-Ukraine war** in 2022 has dramatically reshaped Europe’s approach to fossil fuel dependence, pushing the continent to seek alternatives to Russian gas and oil, while also accelerating efforts to develop renewable energy infrastructure.

#### 6. **How has Europe’s energy mix shifted from domestic coal production to importing fossil fuels from other regions (e.g., Russia, the Middle East)?**

In the mid-20th century, coal was the dominant fuel source, with European countries such as the UK, Germany, and Poland being significant producers. However, the decline of coal in Europe, due to environmental concerns and the EU’s decarbonization goals, has led to increased reliance on natural gas and oil imports, particularly from Russia and the Middle East. This transition has been further accelerated by geopolitical tensions, such as the OPEC oil embargo in the 1970s, the collapse of the Soviet Union, and more recently, the Russian invasion of Ukraine, which has prompted Europe to seek energy diversification through **LNG imports** and **renewable energy sources**.

```python
# Path to data files
file_path = 'data/ener.csv'
shapefile_path = 'data/maps/ne_110m_admin_0_countries.shp'

# Load energy consumption data
df = pd.read_csv(file_path)

# Load world map
world = gpd.read_file(shapefile_path)

# Filter for European countries based on a list
european_countries = [
    'Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 'Denmark', 'Estonia', 'Finland', 'France',
    'Georgia', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Italy', 'Kazakhstan', 'Kosovo', 'Latvia',
    'Liechtenstein', 'Lithuania', 'Luxembourg', 'Malta', 'Moldova', 'Monaco', 'Montenegro', 'Netherlands', 'North Macedonia',
    'Norway', 'Poland', 'Portugal', 'Romania', 'Russia', 'San Marino', 'Republic of Serbia', 'Slovakia', 'Slovenia', 'Spain',
    'Sweden', 'Switzerland', 'Turkey', 'Ukraine', 'United Kingdom', 'Vatican City', 'Albania', 'Andorra', 'Armenia',
    'Azerbaijan', 'Belarus', 'Bosnia and Herzegovina'
]

# Filter map for European countries
europe = world[world['SOVEREIGNT'].isin(european_countries)].copy()

# List of energy consumption columns
consumption_columns = [
    'Biofuels consumption - TWh', 
    'Solar consumption - TWh', 
    'Wind consumption - TWh', 
    'Hydro consumption - TWh', 
    'Nuclear consumption - TWh', 
    'Gas consumption - TWh', 
    'Coal consumption - TWh', 
    'Oil consumption - TWh'
]

# Fill missing values with zeros
df[consumption_columns] = df[consumption_columns].fillna(0)

# Function to plot the map
def plot_map(selected_year):
    plt.clf()  # Clear previous plot

    # Filter data for the selected year
    data_year = df[df['Year'] == selected_year].copy()

    # Ensure rows with zero total consumption don't cause NaN
    data_year['total_consumption'] = data_year[consumption_columns].sum(axis=1)
    data_year = data_year[data_year['total_consumption'] > 0]
    data_year.loc[:, consumption_columns] = data_year[consumption_columns].div(data_year['total_consumption'], axis=0)

    # Calculate cosine similarity matrix
    similarity_matrix = cosine_similarity(data_year[consumption_columns].values)

    # Classification using KMeans clustering
    num_clusters = 8  # Number of clusters (adjust as needed)
    kmeans = KMeans(n_clusters=num_clusters, random_state=0)
    clusters = kmeans.fit_predict(similarity_matrix)

    # Add cluster to data
    data_year['Cluster'] = clusters

    # Mapping cluster to colors - will be adjusted based on the dominant fuel
    data_year['Dominant_Fuel'] = data_year[consumption_columns].idxmax(axis=1)

    # Create a color map for the dominant fuel
    fuel_colors = {
        'Biofuels consumption - TWh': 'green',
        'Solar consumption - TWh': 'yellow',
        'Wind consumption - TWh': 'blue',
        'Hydro consumption - TWh': 'lightblue',
        'Nuclear consumption - TWh': 'orange',
        'Gas consumption - TWh': 'red',
        'Coal consumption - TWh': 'brown',
        'Oil consumption - TWh': 'black'
    }

    data_year['Dominant_Fuel_Color'] = data_year['Dominant_Fuel'].map(fuel_colors)

    # Set default color to gray for countries without a dominant consumption value
    data_year['Dominant_Fuel_Color'] = data_year['Dominant_Fuel_Color'].fillna('gray')

    # Add colors to the map
    europe['Dominant_Fuel'] = europe['SOVEREIGNT'].map(lambda country: data_year[data_year['Entity'] == country]['Dominant_Fuel'].values[0] if country in data_year['Entity'].values else 'None')
    europe['Dominant_Fuel_Color'] = europe['Dominant_Fuel'].map(fuel_colors)

    # Set default color for 'None' values to gray
    europe['Dominant_Fuel_Color'] = europe['Dominant_Fuel_Color'].fillna('gray')

    # Plot the map
    fig, ax = plt.subplots(1, 1, figsize=(15, 10))
    europe.plot(ax=ax, color=europe['Dominant_Fuel_Color'], edgecolor='black')

    # Title of the map
    plt.title(f"Energy consumption similarity in {selected_year} (Dominant energy consumption)")

    # Add legend for dominant energy consumption
    handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=label) 
               for label, color in fuel_colors.items()]
    ax.legend(handles=handles, title="Dominant energy consumption type", bbox_to_anchor=(1.05, 1), loc='upper left')
    ax.set_xlim(-15.5, 40.5)  # Latitude: 12.5 to 14.5 (for Austria)
    ax.set_ylim(26.5, 80.5)  # Longitude: 46.5 to 48.5 (for Austria)

    # Add explanation to the map
    explanation_text = (
        f"This map shows the dominant types of energy fuels consumed by each country in {selected_year}. "
        "Countries are grouped based on the share of different types of energy consumed (e.g. bioenergy, solar, wind, gas). "
        "Each color on the map represents the dominant energy type for the given country."
    )
    
    # Place explanatory text
    plt.figtext(0.5, 0.02, explanation_text, wrap=True, horizontalalignment='center', fontsize=12, color='black')

    # Display the map
    plt.axis('off')
    plt.show()

# Widget for year selection
year_slider = widgets.IntSlider(
    value=2000,
    min=1970,
    max=2023,
    step=1,
    description='Year:',
    continuous_update=False  # This disables immediate updates, only when the user releases the slider
)

# Interactive plot using the slider
interactive_plot = widgets.interactive(plot_map, selected_year=year_slider)

# Display the slider and interactive plot
display(interactive_plot)
```

### Answer: Europe's Dependence on Fossil Fuels

The code addresses the question of Europe's dependence on fossil fuels (oil, coal, and natural gas) by visualizing the energy consumption patterns across European countries. Here's how it works:

- **Energy Mix Visualization**: The map generated by the code shows the dominant fuel type (oil, coal, gas, or renewables) in each country’s energy consumption for a given year. The colors represent different fuel types, with fossil fuels like oil (pink), coal (brown), and gas (dark red) clearly highlighted.

- **Proportional Color Mixing**: The code blends colors based on the share of each fuel type in the total energy consumption. This allows for a more accurate representation of countries that use a mix of fossil fuels and renewables, with a weighted color indicating the energy balance.

- **Renewable Energy Highlight**: Countries with a higher share of renewable energy (biofuels, solar, wind, hydro) are emphasized with more intense green or bright colors. This contrasts with countries more dependent on fossil fuels, which show darker colors.

In summary, the code provides a visual representation that helps to identify which European countries are more reliant on fossil fuels and how this dependency has changed over time.

```python
# Path to data files
file_path = 'data/ener.csv'
shapefile_path = 'data/maps/ne_110m_admin_0_countries.shp'

# Load energy consumption data
df = pd.read_csv(file_path)

# Load world map
world = gpd.read_file(shapefile_path)

# Filter to European countries by list
european_countries = [
    'Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 'Denmark', 'Estonia', 'Finland', 'France',
    'Georgia', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Italy', 'Kazakhstan', 'Kosovo', 'Latvia',
    'Liechtenstein', 'Lithuania', 'Luxembourg', 'Malta', 'Moldova', 'Monaco', 'Montenegro', 'Netherlands', 'North Macedonia',
    'Norway', 'Poland', 'Portugal', 'Romania', 'Russia', 'San Marino', 'Republic of Serbia', 'Slovakia', 'Slovenia', 'Spain',
    'Sweden', 'Switzerland', 'Turkey', 'Ukraine', 'United Kingdom', 'Vatican City', 'Albania', 'Andorra', 'Armenia',
    'Azerbaijan', 'Belarus', 'Bosnia and Herzegovina'
]

# Filter the map to European countries
europe = world[world['SOVEREIGNT'].isin(european_countries)].copy()

# List of energy consumption columns
consumption_columns = [
    'Biofuels consumption - TWh', 
    'Solar consumption - TWh', 
    'Wind consumption - TWh', 
    'Hydro consumption - TWh', 
    'Nuclear consumption - TWh', 
    'Gas consumption - TWh', 
    'Coal consumption - TWh', 
    'Oil consumption - TWh'
]

# Fill missing values with zeros
df[consumption_columns] = df[consumption_columns].fillna(0)

# Function to mix colors based on the share of fuel consumption
def mix_colors(fuel_proportions, fuel_colors):
    # Mixing colors based on the proportion of each fuel
    mixed_color = np.zeros(3)
    
    for i, proportion in enumerate(fuel_proportions):
        # Convert color to RGB
        color_rgb = np.array(mcolors.to_rgb(fuel_colors[i]))
        mixed_color += proportion * color_rgb  # Weighted addition of color based on the proportion
    
    return mixed_color

# Function to plot the map
def plot_map(selected_year):
    plt.clf()  # Clear previous plot

    # Filter data for the selected year
    data_year = df[df['Year'] == selected_year].copy()

    # Ensure that rows with zero total consumption do not cause NaN
    data_year['total_consumption'] = data_year[consumption_columns].sum(axis=1)
    data_year = data_year[data_year['total_consumption'] > 0]
    data_year.loc[:, consumption_columns] = data_year[consumption_columns].div(data_year['total_consumption'], axis=0)

    # Calculate the cosine similarity matrix
    similarity_matrix = cosine_similarity(data_year[consumption_columns].values)

    # Classification using KMeans clustering
    num_clusters = 8  # Number of clusters (adjust as needed)
    kmeans = KMeans(n_clusters=num_clusters, random_state=0)
    clusters = kmeans.fit_predict(similarity_matrix)

    # Add cluster information to the data
    data_year['Cluster'] = clusters

    # Mapping colors for each fuel type
    fuel_colors = [
        'darkgreen',  # Biofuels (dark green)
        'lightgreen', # Solar (light green)
        'mediumblue', # Wind (medium blue)
        'darkblue',  # Hydro (dark blue)
        'yellow',     # Nuclear (yellow)
        'darkred',    # Gas (dark red)
        'brown',      # Coal (brown)
        'pink'        # Oil (pink)
    ]
    
    # Calculate fuel shares and mix colors for each country
    mixed_colors = []
    for _, row in data_year.iterrows():
        # Sort fuels by proportion and take the top three
        sorted_fuels = row[consumption_columns].sort_values(ascending=False).head(3)
        fuel_proportions = sorted_fuels.values
        fuel_names = sorted_fuels.index

        # Mix colors based on the top three fuels
        color = mix_colors(fuel_proportions, [fuel_colors[consumption_columns.index(name)] for name in fuel_names])
        
        # Highlight renewable sources
        renewable_share = row['Biofuels consumption - TWh'] + row['Solar consumption - TWh'] + row['Wind consumption - TWh'] + row['Hydro consumption - TWh']
        if renewable_share / row['total_consumption'] > 0.05:
            color = np.clip(color * 1.2, 0, 1)  # Enhance for renewable sources (more than 50% renewable energy)
        
        mixed_colors.append(color)
    
    # Add mixed colors to the data
    data_year['Mixed_Color'] = mixed_colors

    # Add mixed colors to the map
    europe['Mixed_Color'] = europe['SOVEREIGNT'].map(lambda country: data_year[data_year['Entity'] == country]['Mixed_Color'].values[0] if country in data_year['Entity'].values else [0.5, 0.5, 0.5])  # Default grey if no data

    # Plot the map
    fig, ax = plt.subplots(1, 1, figsize=(15, 10))
    europe.plot(ax=ax, color=europe['Mixed_Color'], edgecolor='black')

    # Map title
    plt.title(f"Energy Consumption Similarity in {selected_year} (Main Energy Consumption Types)")

    # Add legend for dominant energy consumption types
    handles = [
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='darkgreen', markersize=10, label='Biofuels'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='lightgreen', markersize=10, label='Solar'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='mediumblue', markersize=10, label='Wind'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='darkblue', markersize=10, label='Hydro'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='yellow', markersize=10, label='Nuclear'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='darkred', markersize=10, label='Gas'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='brown', markersize=10, label='Coal'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='pink', markersize=10, label='Oil')
    ]
    
    ax.legend(handles=handles, title="Types of Energy Consumption", bbox_to_anchor=(1.05, 1), loc='upper left')
    ax.set_xlim(-15.5, 40.5)
    ax.set_ylim(26.5, 80.5)

    # Add explanation text to the map
    explanation_text = (
        f"This map shows the dominant fuel types in the energy consumption of individual countries in {selected_year}. "
        "Countries are grouped by the share of different types of energy consumed (e.g., bioenergy, solar, wind, gas). "
        "Countries with a higher share of renewable energy sources (bioenergy, solar, wind, hydro) have more intense green colors."
    )
    
    # Position explanation text
    plt.figtext(0.5, 0.02, explanation_text, wrap=True, horizontalalignment='center', fontsize=12, color='black')

    # Display the map
    plt.axis('off')
    plt.show()

# Widget for year selection
year_slider = widgets.IntSlider(
    value=2000,
    min=1970,
    max=2023,
    step=1,
    description='Year:',
    continuous_update=False
)

# Interactive plot using the slider
interactive_plot = widgets.interactive(plot_map, selected_year=year_slider)

# Display the slider and interactive plot
display(interactive_plot)
```

```python
# Path to data files
file_path = 'data/Europ.csv'
shapefile_path = 'data/maps/ne_110m_admin_0_countries.shp'

# Loading energy consumption data
df = pd.read_csv(file_path)

# Loading the world map
world = gpd.read_file(shapefile_path)

# Filtering for European countries based on the list
european_countries = [
    'Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 'Denmark', 'Estonia', 'Finland', 'France',
    'Georgia', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Italy', 'Kazakhstan', 'Kosovo', 'Latvia',
    'Liechtenstein', 'Lithuania', 'Luxembourg', 'Malta', 'Moldova', 'Monaco', 'Montenegro', 'Netherlands', 'North Macedonia',
    'Norway', 'Poland', 'Portugal', 'Romania', 'Russia', 'San Marino', 'Republic of Serbia', 'Slovakia', 'Slovenia', 'Spain',
    'Sweden', 'Switzerland', 'Turkey', 'Ukraine', 'United Kingdom', 'Vatican City', 'Albania', 'Andorra', 'Armenia',
    'Azerbaijan', 'Belarus', 'Bosnia and Herzegovina'
]
# Filtering the map for European countries
europe = world[world['SOVEREIGNT'].isin(european_countries)].copy()

# List of energy consumption columns
consumption_columns = [
    'Biofuels consumption - TWh', 
    'Solar consumption - TWh', 
    'Wind consumption - TWh', 
    'Hydro consumption - TWh', 
    'Nuclear consumption - TWh', 
    'Gas consumption - TWh', 
    'Coal consumption - TWh', 
    'Oil consumption - TWh'
]

# Filling missing values with zeros
df[consumption_columns] = df[consumption_columns].fillna(0)

# Function to plot the map
def plot_map(selected_year):
    plt.clf()  # Clear the previous plot

    # Filter data for the selected year
    data_year = df[df['Year'] == selected_year].copy()

    # Ensure rows with zero total consumption do not cause NaN
    data_year['total_consumption'] = data_year[consumption_columns].sum(axis=1)
    data_year = data_year[data_year['total_consumption'] > 0]
    
    # Calculate the share of renewable and clean sources
    renewable_columns = ['Biofuels consumption - TWh', 'Solar consumption - TWh', 'Wind consumption - TWh', 'Hydro consumption - TWh', 'Nuclear consumption - TWh']
    data_year['renewable_share'] = data_year[renewable_columns].sum(axis=1) / data_year['total_consumption']

    # Map the renewable share to colors
    norm = plt.Normalize(vmin=0, vmax=1)  # Normalize the share between 0 and 1
    cmap = plt.cm.YlGn  # Use the yellow-green color palette

    # Add renewable share to the map
    europe['renewable_share'] = europe['SOVEREIGNT'].map(
        lambda country: data_year[data_year['Entity'] == country]['renewable_share'].values[0] if country in data_year['Entity'].values else 0
    )
    
    # Map the colors according to the renewable share
    europe['Color'] = europe['renewable_share'].map(lambda x: cmap(norm(x)))

    # Plot the map
    fig, ax = plt.subplots(1, 1, figsize=(15, 10))
    europe.plot(ax=ax, color=europe['Color'], edgecolor='black')

    ax.set_xlim(-15.5, 40.5)
    ax.set_ylim(26.5, 80.5)
    # Title of the map
    plt.title(f"Share of renewable and clean sources in the energy mix in {selected_year}")

    # Add the legend
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])  # Needed for the legend
    cbar = plt.colorbar(sm, ax=ax, orientation='vertical', fraction=0.02, pad=0.04)
    cbar.set_label('Share of renewable and clean sources')

    # Add an explanation to the map
    explanation_text = (
        f"This map shows the share of renewable and clean sources in the energy mix of each country in {selected_year}. "
        "Countries are colored according to this share, with dark green indicating a high share and light green indicating a low share."
    )
    
    # Position the explanation text
    plt.figtext(0.5, 0.02, explanation_text, wrap=True, horizontalalignment='center', fontsize=12, color='black')

    # Show the map
    plt.axis('off')
    plt.show()

# Widget for selecting the year
year_slider = widgets.IntSlider(
    value=2000,
    min=1970,
    max=2023,
    step=1,
    description='Year:',
    continuous_update=False  # This disables immediate updates, only when the user releases the slider
)

# Interactive plotting of the map using the slider
interactive_plot = widgets.interactive(plot_map, selected_year=year_slider)

# Display the slider and the interactive plot
display(interactive_plot)
```

### 1. **How has Europe's energy mix developed over time?**

Europe's energy mix has undergone significant transformation over the decades, largely influenced by technological advancements, geopolitical events, and environmental policies. Over time:

- **Pre-1970s:** Europe had a high reliance on domestic fossil fuels, including coal, oil, and natural gas. Coal was especially dominant, with countries like the UK, Germany, and Poland being major producers.
  
- **1970s:** The **1973 oil crisis** and the **1979 energy crisis** marked a pivotal turning point. These crises underscored the risks of over-reliance on foreign oil, especially from the Middle East. As a result, European countries began exploring energy diversification and boosting domestic renewable sources like hydroelectric power.

- **1980s-1990s:** During this period, **nuclear power** emerged as a significant alternative to fossil fuels, particularly in France, which began heavily investing in nuclear energy. However, Europe remained largely dependent on coal and oil, with gas also becoming more prevalent, especially after the collapse of the Soviet Union and the establishment of strong natural gas trade links with Russia.

- **2000s-2010s:** The shift towards **renewable energy sources** (solar, wind, and biofuels) gained momentum, driven by the European Union's ambitious climate goals and the need for energy security. While natural gas and nuclear remained integral parts of the energy mix, there was a notable shift away from coal, especially in the EU-15 countries, due to environmental concerns and regulations.

- **2020s:** The global **energy crisis**, exacerbated by the **Russia-Ukraine War** and supply chain disruptions during the **COVID-19 pandemic**, has accelerated Europe’s transition to alternative energy sources, including renewables, and **liquefied natural gas (LNG)** imports from the US and other non-Russian sources. However, Europe still faces challenges in fully replacing fossil fuels in the short term due to supply and infrastructure limitations.

Overall, Europe has significantly reduced its reliance on domestic coal while increasing the share of **renewable energy** and **imported natural gas**, especially from the Middle East and Russia, though geopolitical tensions have prompted a reevaluation of these energy ties.

### 2. **What role have major geopolitical events played in shaping Europe's fossil fuel dependence?**

Major geopolitical events have had a profound impact on Europe’s energy landscape, driving policy changes, shifts in energy imports, and adjustments in fossil fuel dependence:

- **1973 Oil Crisis:** The **OPEC oil embargo** highlighted Europe’s vulnerability to external energy supplies, particularly from the Middle East. This event forced Europe to reassess its energy security and led to efforts to diversify energy sources. The crisis triggered a push towards alternative energy development and increased exploration of nuclear and renewable options, while also spurring efforts to reduce energy consumption.

- **The Fall of the Soviet Union (1991):** The dissolution of the Soviet Union and the subsequent **opening of Russian energy exports** to Europe significantly altered Europe's energy mix. Russia became a major supplier of natural gas, with pipeline infrastructure developed in the 1990s and 2000s, notably through projects like **Nord Stream**. This dependency on Russian gas grew, though it has now been reconsidered due to geopolitical tensions, particularly following the 2022 Russia-Ukraine war.

- **Russia-Ukraine War (2022):** The invasion of Ukraine has accelerated Europe's efforts to reduce its dependence on Russian fossil fuels, with **sanctions** and a push towards diversifying energy sources. The war disrupted Russia's ability to supply natural gas and oil to Europe, leading to **higher energy prices** and prompting Europe to seek alternative supplies, including **LNG imports** from the US, Qatar, and other nations.

- **Global Financial Crisis (2008):** The financial crisis brought attention to energy costs and the economic impact of high oil prices, further influencing Europe’s push for **energy efficiency** and the expansion of renewable energy sectors.

- **COVID-19 Pandemic (2020):** The pandemic caused massive disruptions to global energy markets, but it also reinforced the importance of **energy resilience** and **local energy production**, pushing Europe towards greater investment in renewable sources to ensure energy security.

### 3. **How has Europe’s energy mix shifted from domestic coal production to importing fossil fuels from other regions (e.g., Russia, the Middle East)?**

Historically, **domestic coal production** played a central role in Europe's energy mix, particularly in countries like the UK, Germany, and Poland. However, several factors have driven a shift from coal to imported fossil fuels:

- **Coal Decline:** Due to environmental concerns, stricter emissions regulations, and the cost of maintaining coal infrastructure, coal's role in Europe’s energy mix has significantly decreased over the last few decades. The EU's commitment to reducing carbon emissions and the **Paris Agreement** further accelerated the phase-out of coal, particularly in the power generation sector.

- **Shift to Natural Gas:** As coal declined, Europe began importing **natural gas** from regions like **Russia** and the **Middle East**, as natural gas was considered a cleaner alternative to coal for electricity generation. The development of **natural gas pipelines** from Russia (e.g., **Yamal-Europe**, **Nord Stream**) increased Europe's dependency on Russian gas. This dependency grew over time, particularly in the 2000s and 2010s.

- **Renewable Energy Expansion:** The 2000s saw the growth of **wind, solar, and bioenergy** across Europe. Countries like Denmark and Germany became leaders in wind energy, and by the 2010s, renewable energy sources began to replace coal and imported oil in many countries.

- **Energy Imports from the Middle East:** In addition to Russian gas, Europe has long imported oil from the **Middle East**. Crude oil prices have had a significant influence on European energy markets, with the **1973 oil crisis** underscoring Europe's reliance on Middle Eastern oil. While oil remains a significant component of Europe’s energy mix, the role of oil has decreased with the rise of renewable energy.

- **Post-Ukraine War:** The **Russia-Ukraine War** has accelerated Europe's move away from Russian energy imports, particularly gas. Countries like Germany, Poland, and others are reducing their reliance on Russian gas and turning towards **LNG imports** from other regions like the **US**, **Qatar**, and **Norway**, as well as ramping up domestic **renewable energy production**.

Overall, Europe has made a major shift from relying heavily on domestic coal to importing fossil fuels, particularly natural gas from Russia and oil from the Middle East. This transition has been driven by a combination of geopolitical events, environmental policies, and technological advancements. The latest challenge is how to navigate energy security, especially in light of ongoing geopolitical tensions and the urgent need to decarbonize the energy sector.
